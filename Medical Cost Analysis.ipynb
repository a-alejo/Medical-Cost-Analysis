# Insurance Analysis Data Cleaning And Visualization

```python
# Load libraries and the dataset - Preview data with number of rows/columns

import pandas as pd
import os
cwd = os.getcwd()

df = pd.read_csv(cwd + '/insurance.csv')
df
```

```python
# Check the data types for each column
df.dtypes
```

## Columns from the dataset:
- age: age of primary policyholder
- sex: gender
- bmi: Body Mass Index (BMI), which is a measure of body weight
- children: Number of dependents covered by health insurance
- smoker: Smoking status
- region: the policyholder residential region in the US, categorized as northeast, southeast, southwest, or northwest
- charges: Individual medical costs billed by health insurance

```python
# Finding the mean (average) of the columns
df.mean(numeric_only=True)
```

```python
# Finding for correlation between columns
df.corr(numeric_only=True)
```

```python
# Finding the number of non null values in each column
df.count()
```

```python
# Finding the max value in each column
df.max()
```

```python
# Finding the lowest value in each column
df.min()
```

```python
# Finding the median (middle number) of the columns
df.median(numeric_only=True)
```

```python
# Finding the standard deviation value (how different the values in a group are from each other) in each column
df.std(numeric_only=True)
```

```python
# Finding the amount of null (blanks) in each column
df.isnull().sum()
```

## Visualizing The Data

```python
# Load visualization libraries
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
sns.boxplot(data=df.sort_values('region'), x='region', y='bmi')
plt.title('Region vs BMI')
```

```python
# Set the width and height of the figure
plt.figure(figsize=(15,5))

sns.lineplot(data=df, x='age', y='charges')
plt.title('Age vs Charges')
```

```python
# Set the width and height of the figure
plt.figure(figsize=(15,5))

sns.countplot(data=df, x='age')
plt.title('Age distribution across the dataset')
```

```python
# Set the width and height of the figure
plt.figure(figsize=(15,5))

sns.barplot(data=df, x='sex', y='children')
plt.title('Number of Children by Gender')
```
